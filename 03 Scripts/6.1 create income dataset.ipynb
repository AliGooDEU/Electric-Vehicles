```python
seven_df.head()
```

```python
# Move column '2010 Median Income' two places to the right
column_to_move = '2010 Median Income'
new_position = seven_df.columns.get_loc(column_to_move) + 2  # Current position + 2

# Create new column order
columns = seven_df.columns.tolist()
# Remove the column from its original position
columns.remove(column_to_move)
# Insert the column at the new position
columns.insert(new_position, column_to_move)

# Reassign the columns to the DataFrame
seven_df = seven_df[columns]
```
